```python
def web_search(query:str):
    url="https://api.duckduckgo.com/"
    params={
        "q":query,
        "format":"json"
    }
    response=requests.get(url,params=params)
    data=response.json()
    
    Abstract=data.get("Abstract","no detailed info is found")
    return Abstract
my_researching=web_search("what is indian history")
print(my_researching)
```
